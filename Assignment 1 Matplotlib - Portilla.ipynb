```python
import numpy as np
```

```python
m = np.linspace(0, 10, 11)
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

PART TWO : PLOTTING E=mc^2

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(m, E, color = "blue", lw=3);
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.xlim(0,10);
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(m, E, c="orange", lw=3);
```

PART THREE (BONUS)

```python
plt.plot(m, E, color = "green", lw=3);
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.xlim(0,10)
plt.yscale("log");
plt.grid(which="both", axis="y")
```

TASK TWO : CREATING PLOTS FROM DATA POINTS

PART ONE: YIELD CURVE DATA

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(labels, july16_2007, label = "july16_2007")
axes.plot(labels, july16_2020, label = "july16_2020");
axes.legend();
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(labels, july16_2007, lw = 3, label = "july16_2007")
axes.plot(labels, july16_2020, lw = 3, label = "july16_2020");
axes.legend(loc=(1.01, 0.4));
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (12,8))
axes[0].plot(labels, july16_2007, lw = 3, label = "july16_2007")
axes[0].set_title("july16_2007")
axes[1].plot(labels, july16_2020, lw = 3, label = "july16_2020")
axes[1].set_title("july16_2020");
```

```python
fig, ax1 = plt.subplots(figsize=(12,8))

ax1.plot(labels, july16_2007, lw=3, color = "blue")
ax1.set_ylabel("July 16 2007", fontsize = 20, color = "blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=3, color = "red")
ax2.set_ylabel("July 16 2020", fontsize = 20, color = "red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
fig, ax1 = plt.subplots(figsize=(12,8))

ax1.spines["left"].set_color("blue")
ax1.spines["left"].set_linewidth(2)

ax1.spines["right"].set_color("red")
ax1.spines["right"].set_linewidth(2)

ax1.plot(labels, july16_2007, lw=3, color = "blue")
ax1.set_ylabel("July 16 2007", fontsize = 20, color = "blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=3, color = "red")
ax2.set_ylabel("July 16 2020", fontsize = 20, color = "red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```
